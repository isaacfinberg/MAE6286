```python
import numpy
from matplotlib import pyplot
% matplotlib inline
```

```python
def linear_convection( nx, L = 2., c = 1., sigma = .5, nt = 20 ):
    
    dx = L / ( nx - 1 )
    x = numpy.linspace( 0., L, nx )
    dt = sigma * dx / c
    u0 = numpy.ones( nx )
    mask = numpy.where( ( x >= .5 ) & ( x <= 1. ) )
    u0[ mask ] = 2.
    u = u0.copy()
    
    for n in range( nt ):
        
        un = u.copy()
        u[ 1: ] = un[ 1: ] - c * dt / dx * ( un[ 1: ] - un[ :-1 ] )
        
    pyplot.figure( figsize = ( 4., 4. ) )
    pyplot.grid()
    pyplot.plot( x, u0, label = 'Initial' )
    pyplot.plot( x, u, label = 'Current' )
    pyplot.legend()
    pyplot.xlim( 0., L )
    pyplot.ylim( 0., 2.5 )
    
```

```python
linear_convection( 41 )
```

```python
linear_convection( 61 )
```

```python
linear_convection( 71 )
```

```python
linear_convection( 85 )
```
